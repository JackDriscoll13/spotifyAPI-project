```python
# Function Defintions 

# Searches for all artists a-z
# Returns a dictionary of artists who have a popularity score that is greater than minpopularityscore.
# the dictionary contains releveant information for each artist including genres, number of followers, popularity score, and ID
def getPopularArtistsDict(letList, minPopularityScore):
    topArtistList = []
    topArtistDict = {}
    counter = 0
    
    for let in letList:
        searchResponse = requests.get("https://api.spotify.com/v1/search?q="+let+"&type=artist", headers=headers).json()
        for artist in searchResponse['artists']['items']:
            if artist['popularity'] > minPopularityScore and artist['name'] not in topArtistList:
                counter+=1
                #append artist name to list, append artist relevant info to dict
                topArtistList.append(artist['name'])
                topArtistDict[artist['name']]={'Relevant Info':{"Genres":artist['genres'], "Followers":artist['followers']['total'],"Popularity":artist['popularity'],"ID":artist['id']}}
                
    print("The top "+str(len(topArtistList))+" artists based on popularity score")
    print(topArtistList)
    return topArtistDict


# Decided to scrap this function and just put it into a dataframe using the above dictionary ^^: 
# def getArtistInfoDF(letlist,minPopularityScore):
#     topArtistList = []
#     topArtistDict = {}
#     counter = 0
#     artistDF = genreDF = pd.DataFrame(data=None,dtype=float,index=None,columns=['Artist Name','TestID','Popularity Score','Followers','spotID','Genres'])
#     for let in letList:
#         searchResponse = requests.get("https://api.spotify.com/v1/search?q="+let+"&type=artist", headers=headers).json()
#         for artist in searchResponse['artists']['items']:
#             if artist['popularity'] > minPopularityScore and artist['name'] not in topArtistList:
```

# SPOTIFY GENRE INFORMATION

```python
#Function Defintions


def getGenreInfoDF(letList, minPopularityScore):
    artistCounter = 0
    rowCounter = 0 
    #uniqueGenres = ()
    genreDict = {}
    topArtistList = []
    genreDF = pd.DataFrame(data=None,dtype=float,index=None,columns=['Genre','Artist Name','TestID','Popularity Score','Followers','spotID','Instance Count'])
    for let in letList:
        searchResponse = requests.get("https://api.spotify.com/v1/search?q="+let+"&type=artist", headers=headers).json()
        for artist in searchResponse['artists']['items']:
            if artist['popularity'] > minPopularityScore and artist['name'] not in topArtistList:
                artistCounter +=1
                topArtistList.append(artist['name'])
                for genre in artist['genres']:
                    #uniqueGenres.add
                    # create a counter that counts how many times each genre is counted
                    if genre in genreDict:
                        genreDict[genre]+=1
                    elif genre not in genreDict:
                        genreDict[genre] = 1
                    rowCounter +=1
                    genreDF.loc[rowCounter] = [genre,artist['name'],artistCounter,int(artist['popularity']),int(artist['followers']['total']),artist['id'],1]
                    
    #print(genreDict)
    return genreDF
```

```python
alphabetList = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','0','1']
# Search through tons of artists, save their info if they have a popularity score of 10 or above
genreDF = getGenreInfoDF(alphabetList,10)
genreDF.head(30)
```

```python
# Group by genres

# Return a new data frame with info about the top 300 genres on spotify. 
topGenres = genreDF.groupby("Genre").agg({"Popularity Score":[sum,np.mean],"Followers":[np.mean],"Instance Count":sum}).sort_values(("Popularity Score","sum"),ascending=False).head(300).reset_index()
topGenres
```

```python
# score playlist by percentage
# lets get the top five genres, 
def getTopGenrePercentagesInPlaylistDictionary(sortedDictionary,relevantGenres = 5):
    newDict = {}
    otherCounter = 0 
    for index,genre in enumerate(sortedDictionary):
        if index<relevantGenres:
            newDict[genre] = str(round(sortedDictionary[genre]*100,2)) +"%"
        elif index>relevantGenres:
            otherCounter += sortedDictionary[genre]
                
    newDict["Other"] = str(round(otherCounter*100,2))+"%"
    newDict['TotalNumGenres'] = len(list(sortedDictionary.keys()))
    return newDict 
```

# COACHELLA DATA IMPORT AND UPDATE


```python
# Import Coachella CSV 
coachella2019 = pd.read_csv("Coachella2019.csv")
coachella2019
```

```python


def updateCoachellaDF(coachellaDF):
    populatedDF = pd.DataFrame(data=None,dtype=float,index=None,columns=['Artist','CoachellaRank','SpotCurrentPopularity','SpotFollowers','SpotPrimaryGenre','DayPerformedCoachella','NameMatch?'])
    for i in range(0,len(coachellaDF)):
            artistName = coachellaDF.iloc[i]['ArtistName']
            #print(artistName)
            searchResponse = requests.get("https://api.spotify.com/v1/search?q="+artistName+"&type=artist", headers=headers).json()
            if searchResponse != None and searchResponse['artists']['items']!= []:
                #print(searchResponse['artists']['items'][0]['name'])
                if searchResponse['artists']['items'][0]['name'].lower() == artistName.lower():
                    nameMatch = 'Y'
                else: 
                    nameMatch = 'N'
                artist = searchResponse['artists']['items'][0]['name']
                coachRank = coachellaDF.iloc[i]['Rank']
                popScore = searchResponse['artists']['items'][0]['popularity']
                followers = searchResponse['artists']['items'][0]['followers']['total']
                if searchResponse['artists']['items'][0]['genres'] != []:
                    primaryGenre = searchResponse['artists']['items'][0]['genres'][0]
                else:
                    primaryGenre = 'Genre not found'
                day = coachellaDF.iloc[i]['Day']
                populatedDF.loc[i] = [artist,coachRank,popScore,followers,primaryGenre,day,nameMatch]
    populatedDF          
    return populatedDF
```
