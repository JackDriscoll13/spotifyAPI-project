```python
# Import and settings
import numpy as np 
import pandas as pd 
%matplotlib inline
import seaborn as sb 
import matplotlib.pyplot as plt
pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

## Artist Analysis

```python
# A dataframe of the top 500 artists on spotify (by popularity score)

# Artist DF import and one line of cleaning
topArtistsDF = pd.read_csv("currenttop500ArtistsSpotify")
topArtistsDF = topArtistsDF.drop('Unnamed: 0','columns') # drop unwanted column
topArtistsDF.head(100)
```

```python
# Visualization of artist Name and followers
# Goal: put into perspective how many followers certain artists have -- is there a big discrepency visually?
top40artists = topArtistsDF.sort_values('Followers',ascending=False).head(40)
sb.set_style("whitegrid")
sb.catplot(x='Artist Name',y='Followers',data=top40artists,kind='bar',aspect=7,color='skyblue')
```

```python
#Visualization of artist name and popularity Score
# See the HOT TOP artists of right now
hottest20Artists = topArtistsDF.sort_values('Popularity Score',ascending=False).head(20)
sb.catplot(x='Artist Name',y='Popularity Score',data=hottest20Artists,kind='bar',palette='Reds_r',aspect=5)
```

```python
#Visualization of followers and popularity score --> how closely related are the two?
sb.factorplot(x='Popularity Score',y='Followers',data=topArtistsDF,aspect=3)
```

## Genre Analysis and Visualizations

```python
# A dataframe of the top 300 genres on spotify

# Initial import and some fun cleaning
# .drop and .rename are methods I learned from pandas api refrence.
topGenresDF = pd.read_csv("topGenresSpotify")
topGenresDF = topGenresDF.drop('Unnamed: 0','columns').drop(0,'index') # delete unwanted columns
topGenresDF = topGenresDF.rename(columns ={'Popularity Score':'PopularityScore_Sum','Popularity Score.1':'PopularityScore_Mean','Followers':'Followers_Mean'})
topGenresDF['Followers_Mean'] = topGenresDF['Followers_Mean'].apply(lambda x: round(float(x),2))  #round some of my columns to look neater
topGenresDF['PopularityScore_Mean'] = topGenresDF['PopularityScore_Mean'].apply(lambda x: round(float(x),2))
topGenresDF
```

```python
# Visualize the Genre name and Popularity Score Sum
# is there a big visual discrepency?
top30genres = topGenresDF.head(30)
top30genres['PopularityScore_Sum'] = top30genres['PopularityScore_Sum'].astype(float)
sb.catplot(x='Genre',y='PopularityScore_Sum',data=top30genres,kind='bar',aspect=4, color='purple')
```

```python
# Which genres have the highest MEAN popularity score?
top30genres = topGenresDF.sort_values('PopularityScore_Mean',ascending = False).head(15)
sb.catplot(x='Genre',y='PopularityScore_Mean',data=top30genres,kind='bar',aspect=4, palette='autumn')
```

```python
# Are mean popularity and popularity sum related?
specialGenresDF = topGenresDF
specialGenresDF['PopularityScore_Sum'] = topGenresDF['PopularityScore_Sum'].astype(float)
sb.lineplot(y=topGenresDF['PopularityScore_Sum'],x=topGenresDF['PopularityScore_Mean'])
```
